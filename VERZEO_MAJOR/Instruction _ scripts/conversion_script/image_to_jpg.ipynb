```python
import os
import cv2
```

```python
start_sequence = int(input('From where you want to start the sequence ? '))  #start sequence
```

```python
for count, filename in enumerate(os.listdir("convert_test/")): #listdir takes the directory where unchanged images are
    image = cv2.imread('convert_test/'+filename)   # converted test should be changed to the unchanged directory as in above line.
    cv2.imwrite("converted_jpg/other-"+str(count+start_sequence)+".jpg", image, [int(cv2.IMWRITE_JPEG_QUALITY), 100]) # converted_jpg/ is another directory where converted images will be stored.
```
