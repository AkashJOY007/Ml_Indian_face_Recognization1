### trained model with 76 % accuracy

```python
model = joblib.load('human_other_model_svm(1).pkl')
```

#### paths

```python
path_data = 'test_data/'
path_human = 'seperated/human/'
path_other = 'seperated/other/'
```

```python
def path_exist(path_data):
    if os.path.exists(path_data):
        shutil.rmtree(path_data)
```

```python
path_exist(path_other)
os.mkdir(path_other)
path_exist(path_human)
os.mkdir(path_human)
```

### Seperation using trained model

```python
for count, filename in enumerate(os.listdir(path_data)):
    img = cv2.imread(path_data+filename)
    try:
        scaled_img = cv2.resize(img, (50, 50))
        reshape = scaled_img.reshape(50*50*3,1)
        test_image = np.array(reshape).reshape(1,len(reshape)).astype(float)
        classifier = model.predict(test_image)
        if classifier[0] == 0:
            location = path_human + filename
            cv2.imwrite(location,img)
        else:
            location = path_other + filename
            cv2.imwrite(location,img)
    except Exception as e:
        pass
```
