```python
from sklearn.model_selection import train_test_split
```

```python
X = ds.drop(['Date','Cured','State/UnionTerritory','ConfirmedIndianNational','ConfirmedForeignNational'],axis = 1)
Y = ds['Cured']
```

```python
ds.dtypes
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
print(X)
```

```python
from sklearn.linear_model import RidgeCV
model = RidgeCV(cv=2)
model.fit(X_train, Y_train)
```

```python
predict = model.predict(X_test)
print("{}".format(np.linalg.norm(predict - Y_test, 1)*2/len(Y_test)))
```
